```python
import pandas as pd
import numpy as np
```

```python
#Bir dataframe oluşturalım
df=pd.DataFrame({'A':[1,2,3,4,5,np.nan,7,np.nan,9,np.nan],
                 'B':[10,20,np.nan,30,40,50,np.nan,70,80,np.nan]}
                )
```

```python
df
```

```python
#her bir işlem için datframemimde boşluklar dolacağı için dataframemimin kopyasını oluşturuyorum.
df1=df.copy()
```

```python
df1.info() #dataframe.info ile kolonlarımın veri tipleri ve boş satır bilgilerimi öğreniyorum.
```

```python
df1.isnull() #satır bazlı boş değerlerime bakıyorum True olanlar boş değerler oluyor.
```

```python
df1.isnull().sum() #burada da her bir kolonumda kaç tane boş değer var ona baktım
```

# Ortalama ile boş verileri doldurma

```python
df['A'].mean() #A kolonumuzun dolu olan değerlerinin ortalamasına bakıyorum ki boş verilerimi onunla doldurulacağı için
```

```python
df1['A']=df1['A'].fillna(df1['A'].mean())
```

```python
df1['A']
```

```python
df1
```

```python
df1['B'].mean() #B kolonumuzun dolu olan değerlerinin ortalamasına bakıyorum ki boş verilerimi onunla doldurulacağı için
```

```python
df1['B']=df1['B'].fillna(df1['B'].mean())
```

```python
df1
```

# Medyan ile boş verilerin doldurulması

```python
df2=df.copy() #df1 in boş verilerini doldurduğum için yeni bir df kopyamı oluşturuyorum.
```

```python
df2
```

```python
df2['A'].median() #A kolonumuzun medyanına baktım onunla dolduracağım için
```

```python
df2['A']=df2['A'].fillna(df2['A'].median())
```

```python
df2['A']
```

```python
df2['B'].median()
```

```python
df2['B']=df2['B'].fillna(df2['B'].median())
```

```python
df2['B']
```

#  Eksik verilerin bir önceki ile doldurulması

```python
df3=df.copy()
df3 #yeniden bir dataframe kopyası oluşturuyorum.
```

```python
df3['A']=df3['A'].fillna(method='ffill')
df3['A']
```

```python
df3['B']=df3['B'].fillna(method='ffill')
df3['B']
```

```python
df3
```

# **Bir Sonraki ile Doldurma**

```python
df4=df.copy()
df4 #dataframe kopyası oluşturuyoruz yeniden
```

```python
df4['A']=df4['A'].fillna(method='bfill')
df4['A']
```

```python
df4['B']=df4['B'].fillna(method='bfill')
df4['B']
```

```python
df4
```

Gördüğünüz gibi bir sonraki değerle doldurduğumuzda dataframin son rowları hala nan kaldı bunu düzeltmemiz için aynı zamanda bir önceki ile doldurma uygulamamız lazım

```python
df4.fillna(method='ffill',inplace=True)
```

```python
df4
```

Gördüğünüz gibi bir önceki ile doldurma metodunu uyguladığımızda doldurulmuş oldu

# **SimpleImputer ile doldurma**

```python
df5=df.copy()
df5 #dataframe kopyası oluşturuyoruz
```

```python
from sklearn.impute import SimpleImputer
```

### Method: ortalama ile

```python
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
df5['A']=imp.fit_transform(df5[['A']])
df5['A']
```

```python
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
df5['B']=imp.fit_transform(df5[['B']])
df5['B']
```

```python
df5
```

```python
data = pd.DataFrame({
    'A': ["Istanbul", "Ankara", "Istanbul", "Izmir", "Ankara", "Istanbul", np.nan, "Bartın", "Istanbul", np.nan, "Elazığ", np.nan],
    'B': np.random.randint(0, 100, size=12)
})



# SimpleImputer ile eksik değerleri en sık görülen değerle doldurma
imp = SimpleImputer(strategy='most_frequent')
# fit_transform 2B bir dizi döndürür, bu nedenle bunu 1B'ye düzleştirmemiz gerekir
data['A'] = imp.fit_transform(data[['A']]).ravel()

data
```
