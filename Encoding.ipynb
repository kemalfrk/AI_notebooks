# Merhabalar bugün en çok uygulanan encoding metodlarını uygulamalı bir şekilde göstereceğim

```python
import pandas as pd
import numpy as np
```

```python
data = {'Employee id': [10, 20, 15, 25, 30,50,60],
        'Gender': ['M', 'F', 'F', 'M', 'F','M','F'],
        'Remarks': ['Good', 'Nice', 'Good', 'Great', 'Nice','Good','Nice'],
        }
#pandas dataframine çevirdik
df = pd.DataFrame(data)
df
```

# get_dummy()

```python
df1=df.copy() #üstünde işlemler yapacağım için orijinal verisetimin kopyasını oluşturup onunla işlem yapacağım
```

```python
df1_encoded=pd.get_dummies(df1,columns=['Gender','Remarks'],dtype=int) #burada get_dummies metoduyla kategorik değişkenlere one hot encoding uyguladık
#parametre olarak;hangi kolonlara olacağını seçmek için columns parametresini kullandık,dype=int yaptık çünkü default değerinde bıraksaydık bool veri tipine çeviriyordu(True,False yani)

df1_encoded
```

Gördüğünüz gibi get_dummies() metodumuz kategorik değişkenlerimize one hot encoding uyguladı ve encode ettiğimiz kategorik veri tipli kolonlarımızı silerek get dummy tuzağı dediğimiz tuzaktan kurtuldu (asıl kolonlarımızı silerek karışıklığı önledi).Bu get_dummies metodumuzun en önemli özelliği

# One-Hot-Encoder

```python
df2=df.copy() #üstünde işlemler yapacağım için orijinal verisetimin kopyasını oluşturup onunla işlem yapacağım
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_columns = df.select_dtypes(include=['object']).columns.tolist() #one hot encoding uygulayacağımız kategorik değişkenlerin kolon adlarını liste şeklinde tuttuk
encoder = OneHotEncoder(sparse_output=False) #OneHotEncoder objesi oluşturduk

one_hot_encoded = encoder.fit_transform(df[categorical_columns]) #kategorik değişkenlere one hot encoding uyguladık

one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns)) #uygulanan one hot encoding işleminin ardından bunu dataframe e çeviriyoruz

one_hot_df

```

# Ordinal Encoding

Ordinal Encoding, kategorik verileri doğal sıralamalarını koruyan sayısal değerlere dönüştürmek için kullanılan bir ön işleme tekniğidir.

Sıralı kodlama, kategoriler arasında anlamlı bir sıralama olduğunu varsayar. Sıra içermeyen [“kırmızı”, “yeşil”, “mavi”] gibi nominal kategoriler için kullanılmamalıdır.

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder() #ordinal encoder objesi oluşturduk

sizes = ["small", "medium", "large"] #sıralı kategorik değişken değerleri listesi oluşturduk

sizes = np.array(sizes).reshape(-1,1) #bu listeyi numpy arrayine çevirdik çünkü fit transform ederken bu şekilde kabul ediyor

encoded = encoder.fit_transform(sizes)

encoded
```

```python
encoder.inverse_transform(encoded) #inverse_transform metoduyla hangi kategorik değişkenlerimize hangi numerik değer verdiğini görmüş olduk
```
